```python
import yfinance as yf
import matplotlib.pyplot as plt

# Создание объекта Ticker для заданной компании
msft = yf.Ticker("AAPL")

# Получение цены закрытия акций за последний год
hist = msft.history(period="1y")

# Рассчет скользящего среднего за 30 дней
hist["SMA_30"] = hist["Close"].rolling(window=30).mean()

# Вывод графика цены закрытия акций и скользящего среднего за последний год
plt.plot(hist["Close"])
plt.plot(hist["SMA_30"])
plt.legend(["Close", "SMA_30"])
plt.show()
```

```python
import yfinance as yf
import pandas as pd

# Создание объектов Ticker для двух компаний
tsla = yf.Ticker("TSLA")
aapl = yf.Ticker("AAPL")

# Получение цены закрытия акций за последний год
tsla_hist = msft.history(period="1y")
aapl_hist = aapl.history(period="1y")

# Объединение данных в одну таблицу
data = pd.concat([tsla_hist["Close"], aapl_hist["Close"]], axis=1)
data.columns = ["TSLA", "AAPL"]

# Рассчет корреляции между ценами акций двух компаний
corr = data.corr()

# Вывод матрицы корреляции
print(corr)
```

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Создание объекта Ticker для заданной компании
aapl = yf.Ticker("AAPL")

# Получение цены закрытия акций за последние 5 лет
hist = aapl.history(period="5y")

# Рассчет процентного изменения цены закрытия акций
hist["Pct_Change"] = hist["Close"].pct_change()

# Вывод графика процентного изменения цены закрытия акций
plt.plot(hist["Pct_Change"])
plt.title("AAPL Percent Change")
plt.show()
```

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Создание объекта Ticker для заданной компании
aapl = yf.Ticker("AAPL")

# Получение цены закрытия акций за последние 5 лет
hist = aapl.history(period="5y")

# Рассчет волатильности цены закрытия акций за 30 дней
hist["Volatility"] = hist["Close"].rolling(window=30).std()

# Вывод графика волатильности цены закрытия акций
plt.plot(hist["Volatility"])
plt.title("AAPL Volatility")
plt.show()
```

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Создание объектов Ticker для заданных компаний
tsla = yf.Ticker("TSLA")
aapl = yf.Ticker("AAPL")

# Получение цены закрытия акций за последние 5 лет
tsla_hist = tsla.history(period="5y")
aapl_hist = aapl.history(period="5y")

# Объединение таблиц цен закрытия акций для обеих компаний
merged = pd.merge(tsla_hist["Close"], aapl_hist["Close"], left_index=True, right_index=True, suffixes=("_TSLA", "_AAPL"))

# Вывод графиков цен закрытия акций обеих компаний
merged.plot()
plt.title("TSLA vs AAPL Close Prices")
plt.show()

# Рассчет коэффициента корреляции между ценами закрытия акций обеих компаний
corr = merged.corr().iloc[0,1]

# Вывод коэффициента корреляции
print("TSLA vs AAPL Correlation: {:.2f}".format(corr))
```

```python
import yfinance as yf
import datetime as dt

# Создание объекта Ticker для заданной компании
aapl = yf.Ticker("AAPL")

# Получение цены закрытия акций с начала года до сегодняшнего дня
start_date = dt.datetime.now().strftime('%Y-01-01')
end_date = dt.datetime.now().strftime('%Y-%m-%d')
aapl_hist = aapl.history(start=start_date, end=end_date)

# Рассчет процентного изменения цены закрытия акций
price_change = (aapl_hist["Close"][-1] - aapl_hist["Close"][0]) / aapl_hist["Close"][0] * 100

# Вывод процентного изменения цены закрытия акций
print("AAPL Price Change Since Beginning of Year: {:.2f}%".format(price_change))
```

```python
import yfinance as yf
import matplotlib.pyplot as plt
import datetime as dt
import pandas as pd

# Создание списка из нескольких компаний
tickers = ["MSFT", "AAPL", "AMZN", "TSLA","FRHC"]

# Получение цены закрытия акций за заданный период времени для каждой компании
start_date = "2021-01-01"
end_date = "2021-12-31"
closing_prices = {}
for ticker in tickers:
    stock = yf.Ticker(ticker)
    hist = stock.history(start=start_date, end=end_date)
    closing_prices[ticker] = hist["Close"]

# Создание DataFrame из цен закрытия акций
df = pd.DataFrame(closing_prices)

# Рассчет дневных доходностей для каждой компании
daily_returns = df.pct_change()

# Рассчет совместных доходностей
weights = [0.25, 0.25, 0.25, 0.25, 0.25] # равные веса для каждой компании
portfolio_returns = (daily_returns * weights).sum(axis=1)

# Построение графика совместных доходностей
plt.plot(portfolio_returns)
plt.title("Portfolio Returns for 2021")
plt.xlabel("Date")
plt.ylabel("Returns")
plt.show()
```

```python
import yfinance as yf
import matplotlib.pyplot as plt

# Задание тикеров двух компаний
ticker1 = "FRHC"
ticker2 = "AAPL"

# Получение цен закрытия за заданный период времени для двух компаний
start_date = "2021-01-01"
end_date = "2021-12-31"
stock1 = yf.Ticker(ticker1)
stock2 = yf.Ticker(ticker2)
stock1_hist = stock1.history(start=start_date, end=end_date)
stock2_hist = stock2.history(start=start_date, end=end_date)

# Создание DataFrame из цен закрытия для двух компаний
df = pd.concat([stock1_hist["Close"], stock2_hist["Close"]], axis=1)
df.columns = [ticker1, ticker2]

# Рассчет дневных доходностей для двух компаний
df_returns = df.pct_change()

# Рассчет корреляции между доходностями двух компаний
corr = df_returns.corr().iloc[0,1]

# Построение scatter plot
plt.scatter(df_returns[ticker1], df_returns[ticker2])
plt.title(f"{ticker1} vs {ticker2} - Correlation: {corr:.2f}")
plt.xlabel(f"{ticker1} Returns")
plt.ylabel(f"{ticker2} Returns")
plt.show()
```
